## 绘图函数：

```python
import pandas as pd
import matplotlib.pyplot as plt

da = pd.read_csv('speed.csv')

print(da.shape)
da = da.iloc[5: 200, :]
print(da.shape)
```

```python
plt.figure(figsize=[10,15])

plt.plot(range(da.shape[0]), da, '-')
leg =['with traci', 'without traci']
plt.legend(leg)
plt.show
```
