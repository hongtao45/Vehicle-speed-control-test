## 绘图函数：

```python
import pandas as pd
import matplotlib.pyplot as plt

da = pd.read_csv('speed.csv')

print(da.shape)
da = da.iloc[5: da.shape[0], :] #　第一个速度很小很小，很奇怪，删掉他
print(da.shape)
```

```python
plt.figure(figsize=[10,5])

plt.plot(range(da.shape[0]), da, '-')
leg =['test1', 'test2']
plt.legend(leg)

plt.show
```
